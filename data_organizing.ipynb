```python
def organize_data(file_name, path_to_new_folder, label):
    label_num =int(label_frame[label_frame['display_name']==label]['index'])
    raw_dataset = tf.data.TFRecordDataset(file_name)
    #print(raw_dataset)
    
    good_datasets= []
    index= 0
    
    for x in raw_dataset:
        #print(x)
        #for each data point in the file open it up and look at what's in it
        example = tf.train.Example()
        example.ParseFromString(x.numpy())
        #now make a dictionary of the features we can use to find the right data points 
        result = {}
        #should be a loop of about 4
        for key, feature in example.features.feature.items():
            kind = feature.WhichOneof('kind')
            result[key] = np.array(getattr(feature, kind).value)
        arr= np.array(result['labels'])
        if label_num in arr:
            print(True)
            place_holder_dataset=tf.data.TFRecordDataset.from_tensors(x)
            print(place_holder_dataset)
            good_datasets.append(place_holder_dataset)
            print(index)
            index = index+1
        
        else: 
            #print(False)
            continue
        
    if(len(good_datasets)>0):
        
        new_dataset=good_datasets[0]
        print(good_datasets[0])
        for i in range(1,len(good_datasets)):
            new_dataset=new_dataset.concatenate(good_datasets[i])
            
        new_filename = os.path.join(path_to_new_folder,'new_dataset'+file_name[-11:-8]+'tfrecord')
        print(new_filename)
        writer = tf.data.experimental.TFRecordWriter(new_filename)
        writer.write(new_dataset)
        
        return new_dataset
            
        
    else: 
        return None
    



    
    
    
    
    
    
    
```
